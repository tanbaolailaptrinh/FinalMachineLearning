```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm mục tiêu (ví dụ: f(x) = x^2)
def objective_function(x):
    return x**2

# Đạo hàm của hàm mục tiêu (f'(x) = 2x)
def gradient(x):
    return 2 * x

# Thuật toán Gradient Descent
def gradient_descent(x, learning_rate, epochs):
    x_values = []
    cost_values = []

    for epoch in range(epochs):
        cost = objective_function(x)

        x_values.append(x)
        cost_values.append(cost)

        # Cập nhật giá trị của x bằng cách di chuyển ngược chiều gradient
        x = x - learning_rate * gradient(x)

    return x_values, cost_values

# Thiết lập các tham số
initial_x = -3.0
learning_rate = 0.1
epochs = 20

# Chạy Gradient Descent
x_optimized_values, cost_values = gradient_descent(initial_x, learning_rate, epochs)

# Vẽ biểu đồ cost qua từng chỉ số
plt.plot(range(1, epochs + 1), cost_values, marker='o', linestyle='-', color='b')
plt.title('Cost over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.show()
```
